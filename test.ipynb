```python
from core.rdvq1ac import VQ
from core.rdvq1ac_noT import VQ_noT
from core.util.evaluate import *
from core.data import *
import numpy as np
from core.util import *
import warnings
warnings.filterwarnings("ignore")
from core.OneGridSC import OneGridSC, load_color
from time import gmtime, strftime, localtime
from core.spatialKM import spatialKM

from core.OneGridSCwhole import OneGridSCwhole

def load_hash(key):
    with open('name.pkl', 'rb') as f:
        d = pickle.load(f)
    return d[key]
root = load_hash('root')
g_hash = load_hash('name')


def run_test(grid, new_lambda, skip):    
    tmp = load(Rtype='kodak', ct=[-1], size=[256,128,64,32,8])
    Y256t, Y128t, Y64t, Y32t, Y8t = tmp[0], tmp[1], tmp[2], tmp[3],tmp[4]
    colort = load_color(root, Y256t)
    if grid >= 3:
        g3 = OneGridSC(grid=3, model_hash=g_hash)
        iY8t = g3.predict(np.zeros_like(Y8t), Y8t, colort, Y256t, new_lambda=new_lambda[3], skip=skip.get(3,{}))
        g3 = None
        if grid >=5:
            g5 = OneGridSC(grid=5, model_hash=g_hash)
            iY32t = g5.predict(iY8t, Y32t, colort, Y256t, new_lambda=new_lambda[5], skip=skip.get(5,{}))
            g5, iY8t = None, None
            if grid >=6:
                g6 = OneGridSC(grid=6, model_hash=g_hash)
                iY64t = g6.predict(iY32t, Y64t, colort, Y256t, new_lambda=new_lambda[6], skip=skip.get(6,{}))
                g6, iY32t = None, None
                if grid>=7:
                    g7 = OneGridSC(grid=7, model_hash=g_hash)
                    iY128t = g7.predict(iY64t, Y128t, colort, Y256t, new_lambda=new_lambda[7], skip=skip.get(7,{}))
                    g7, iY64t = None, None
                    if grid>=8:
                        g8 = OneGridSC(grid=8, model_hash=g_hash+'b')
                        iY256t = g8.predict(iY128t, Y256t, colort, Y256t, new_lambda=new_lambda[8], loadonce=False)
                        g8=[]
                        if grid>=81:
                            g82 = OneGridSCwhole(grid=8, model_hash=g_hash+'2b')
                            iY256t = g82.predict(iY256t, Y256t, colort, Y256t, new_lambda=new_lambda[81])
                            g82=[]
                            if grid >=82:
                                g81 = OneGridSCwhole(grid=8, model_hash=g_hash+'1b')
                                iY256t = g81.predict(iY256t, Y256t, colort, Y256t, new_lambda=new_lambda[82])
                                g81=[]
                        print(MSE(iY256t, Y256t))
```
